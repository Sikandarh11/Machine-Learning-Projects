```python
#Removing Html Tags and urls
import re
def remove_html_tags_urls(text):
  text = re.sub("<.*?>", "", text)
  text = re.sub("http\S+|www\S+|https\S+", "", text)
  return text
df['review'] = df['review'].apply(remove_html_tags_urls)
df.head()
```

```python
#removing punctuations
import string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['review'] = df['review'].apply(remove_punctuation)
df.head()
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()

def correct_spelling(text):
    corrected_text = []
    for word in text.split():
        corrected_word = spell.correction(word)
        # If correction returns None, keep the original word
        if corrected_word is not None:
            corrected_text.append(corrected_word)
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)


def correct_spelling(text):
    corrected_text = [
        spell.correction(word) if spell.correction(word) is not None else word
        for word in text.split()
    ]
    return " ".join(corrected_text)

df['review'] = df['review'].apply(correct_spelling)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    return ' '.join([word for word in words if word.lower() not in stop_words])

df['review'] = df['review'].apply(remove_stopwords)
df.head()
```

```python
#tokenizing
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Only needed the first time

def tokenize_words_nltk(text):
    return word_tokenize(text)

newdf['word_tokens'] = newdf['review'].apply(tokenize_words_nltk)
newdf.head()
```

```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

def lemmatize_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])

newdf['lemmatized_text'] = newdf['review'].apply(lemmatize_spacy)
newdf.head()
```

```python
y = df[['sentiment']].iloc[:20000,: ]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

y = encoder.fit_transform(df[['sentiment']].iloc[:20000,: ])
tfidf_matrix.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(tfidf_matrix,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics
 accuracy_score

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
# Naive Bayes
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Naive Bayes Accuracy:", accuracy_nb)
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accuracy_svm)
print(classification_report(y_test, y_pred_svm))
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Gradient Boosting Accuracy:", accuracy_gb)
print(classification_report(y_test, y_pred_gb))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(use_label_encoder = False ,eval_metric = "logloss")
xgb.fit(X_train, y_train)

# Make predictions
y_pred_xgb = xgb.predict(X_test)

print(classification_report(y_test, y_pred_xgb))
```
